## Diccionario de Variables


# Variables de Identificación y Contexto (ID)

Características básicas para identificar cada viaje, unidad o periodo analizado.

Unidad: Identificador del tractocamión o tráiler (ej. T545, T795).

Cliente / Grupo: Cliente o grupo operativo responsable del viaje.

Revisión / Periodo / Mes: Ventana temporal del análisis.

Origen: Punto donde inicia la operación.

Destino: Punto donde finaliza la operación.

Zona: Clasificación territorial o regional.

# Variables de Desempeño Operativo (OUTPUT)

Indicadores resultantes de la operación diaria; miden productividad y eficiencia del proceso.

Kilometraje Bitácora: Distancia recorrida registrada.

Unidades Colocadas: Número de vehículos asignados/operados.

Duración de Conducción / Recorrido: Tiempo efectivo de manejo según telemetría.

Tiempo de Carga: Tiempo utilizado para maniobras de carga.

Tiempo en Ralentí: Tiempo en marcha sin movimiento (consumo no productivo).

Duración (Estadía): Tiempo de espera o permanencia en puntos de carga/descarga.

# Variables de Eficiencia y Costo (INPUT / MÉTRICA)

Indicadores que miden qué tan eficiente se utilizan los recursos y el costo de operación.

Rendimiento: Eficiencia de combustible (km/L).

Costo por km: Costo unitario del recorrido.

Costo / Combustible: Gasto total en combustible.

Litros Consumidos: Cantidad total de combustible utilizado.

Importe (Gasto) / Concepto (Gasto): Detalle de gastos operativos por categoría.

Clasificación de Servicio / Descripción (Falla): Tipo, categoría o frecuencia de fallas y mantenimientos.

# Variables de Calidad y Servicio (OUTCOME)

Indicadores finales que miden el impacto global del servicio y el valor entregado.

%NS (Nivel de Servicio): Cumplimiento de entregas, disponibilidad o KPI de confiabilidad.

Ingreso: Resultado económico directo asociado al viaje o periodo.

% Combustible: Proporción del costo de combustible respecto al ingreso (eficiencia económica).

## Análisis de Valores Faltantes




```python
missing_percentage = viajes.isnull().sum() / len(viajes) * 100
missing_percentage = missing_percentage[missing_percentage > 0].sort_values(ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x=missing_percentage.index, y=missing_percentage, hue=missing_percentage.index, palette='viridis', legend=False)
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Porcentaje de Valores Faltantes (%)', fontsize=12)
plt.title('Porcentaje de Valores Faltantes por Columna', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Análisis de Cardinalidad


```python
object_cols = viajes.select_dtypes(include='object').columns
cardinality = {}
for col in object_cols:
    cardinality[col] = viajes[col].nunique()

cardinality_series = pd.Series(cardinality).sort_values(ascending=False)
print("Cardinalidad de columnas de tipo 'object':")
print(cardinality_series)
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x=cardinality_series.head(10).index, y=cardinality_series.head(10).values, hue=cardinality_series.head(10).index, palette='viridis', legend=False)
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Número de Valores Únicos', fontsize=12)
plt.title('Cardinalidad de las 10 Columnas "Object" con más Valores Únicos', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Inspección de Outliers Temporales y Geográficos


Primero es importante convertir las columnas de fecha a tipo datetime y visualizar su distribución para identificar posibles outliers temporales. Para columnas con información geográfica ('Ruta', 'Sucursal', 'Nombre Cliente'), examinar los valores únicos y sus frecuencias para detectar posibles outliers o inconsistencias geográficas.


```python
date_cols_to_convert = ['Fecha', 'Fecha Vencimiento', 'Fecha Salida', 'Fecha Llegada']

for col in date_cols_to_convert:
    viajes[col] = pd.to_datetime(viajes[col], errors='coerce')

print("Converted date columns to datetime type.")
print(viajes[date_cols_to_convert].info())
```

```python
date_cols_for_plot = ['Fecha', 'Fecha.1', 'Fecha Vencimiento', 'Fecha Salida', 'Fecha Llegada']

plt.figure(figsize=(15, 10))
for i, col in enumerate(date_cols_for_plot):
    plt.subplot(3, 2, i + 1)
    sns.histplot(viajes[col].dropna(), kde=True, color='skyblue')
    plt.title(f'Distribución Temporal de {col}')
    plt.xlabel('Fecha')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
geo_cols = ['Ruta', 'Sucursal', 'Nombre Cliente']

for col in geo_cols:
    print(f"\n--- Análisis de la columna: {col} ---")
    print("Top 10 valores más frecuentes:")
    print(viajes[col].value_counts().head(10))
    print("\nTop 10 valores menos frecuentes (posibles outliers/inconsistencias):")
    print(viajes[col].value_counts().tail(10))
```

# Findings
* Se identificaron múltiples columnas con más del 80–100% de valores faltantes, como UUID CP, Documentos, Peso Descarga Kgs, Dolly y Remolque 2.

* Variables críticas como Fecha Salida, Fecha Llegada y Peso Kgs presentan entre 24% y 28% de datos faltantes, afectando análisis operativos.

* La columna Diferencia tiene 35% de valores faltantes, lo cual limita validaciones de peso.

* Se detectaron 13 filas duplicadas, que podrían distorsionar cálculos agregados.

* La consistencia temporal está comprometida por NaNs en fechas de salida y llegada, impidiendo mediciones de duración.

* Las columnas Fecha y Fecha.1 muestran distribuciones casi idénticas, lo que sugiere posible redundancia.

* En la columna Ruta existen muchos valores con frecuencia mínima (1 viaje), potencialmente indicando errores o rutas atípicas.

* Nombre Cliente también presenta clientes con ocurrencias extremadamente bajas, lo que podría incluir errores de captura.

* Sucursal muestra una distribución natural sin outliers claros, pero con gran disparidad en carga operativa.

* En conjunto, los problemas más críticos son valores faltantes masivos, duplicados y categorías de baja frecuencia que requieren depuración.

```python
viajes_numeric = viajes.select_dtypes(include=['number']).copy()

print("Shape of 'viajes_numeric' after selecting all numeric columns:")
print(viajes_numeric.shape)
print("First 5 rows of 'viajes_numeric':")
print(viajes_numeric.head())
```

```python
numeric_cols_to_plot = ['SubTotal', 'IVA', 'Total', 'Peso Kgs']

for col in numeric_cols_to_plot:
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Histogram
    sns.histplot(viajes_numeric[col].dropna(), kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Histograma de {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frecuencia')

    # KDE Plot
    sns.kdeplot(viajes_numeric[col].dropna(), ax=axes[1], color='salmon', fill=True)
    axes[1].set_title(f'KDE de {col}')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('Densidad')

    plt.tight_layout()
    plt.show()
```

## Detección de Outliers y Comparación de Distribuciones (Box Plots)



```python
numeric_cols_for_outliers = ['SubTotal', 'Total', 'Peso Kgs', 'Diferencia']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols_for_outliers):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=viajes_numeric[col].dropna(), color='lightgreen')
    plt.title(f'Box Plot de {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

## Relaciones Bivariadas (Scatter Plots)




```python
scatter_plots_data = [
    {'x': 'SubTotal', 'y': 'Total', 'title': 'Relación entre SubTotal y Total'},
    {'x': 'SubTotal', 'y': 'IVA', 'title': 'Relación entre SubTotal e IVA'},
    {'x': 'Peso Kgs', 'y': 'Diferencia', 'title': 'Relación entre Peso Kgs y Diferencia'}
]

for plot_info in scatter_plots_data:
    x_col = plot_info['x']
    y_col = plot_info['y']
    title = plot_info['title']

    # Create a temporary DataFrame for plotting, dropping NaNs locally
    plot_df = viajes_numeric[[x_col, y_col]].dropna()

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=plot_df[x_col], y=plot_df[y_col], alpha=0.6)
    plt.title(title, fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

Relaciones Numéricas (Pair Plots)




```python
selected_numeric_cols = ['SubTotal', 'Total', 'IVA', 'Peso Kgs']

# Create a temporary DataFrame with selected columns and drop NaNs
# This ensures that pairplot only considers rows where all selected columns have non-null values
pair_plot_df = viajes_numeric[selected_numeric_cols].dropna()

# Generate the pair plot
sns.pairplot(pair_plot_df)
plt.suptitle('Pair Plot de Variables Numéricas Clave', y=1.02) # y adjusts the title position
plt.tight_layout()
plt.show()
```

## Matriz de Correlación




```python
selected_numeric_cols = ['SubTotal', 'Total', 'IVA', 'Peso Kgs']

correlation_df = viajes_numeric[selected_numeric_cols].dropna()

correlation_matrix = correlation_df.corr(method='pearson')

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación de Variables Numéricas Seleccionadas', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Análisis de Componentes Principales - PCA




```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select the key numerical columns for PCA
pca_data = viajes_numeric[['SubTotal', 'Total', 'IVA', 'Peso Kgs']]

# 2. Remove any rows with null values
pca_data_cleaned = pca_data.dropna()

# 3. Initialize a StandardScaler
scaler = StandardScaler()

# 4. Fit and transform the cleaned data
scaled_data = scaler.fit_transform(pca_data_cleaned)

# 5. Initialize a PCA object and fit it to scaled_data
pca = PCA()
pca.fit(scaled_data)
```

```python
pca_components = pca.transform(scaled_data)
pca_result_df = pd.DataFrame(data = pca_components[:,0:2], columns = ['PC1', 'PC2'])

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', data=pca_result_df, alpha=0.6)
plt.title('Primeros Dos Componentes Principales (PC1 vs PC2)')
plt.xlabel('Primer Componente Principal (PC1)')
plt.ylabel('Segundo Componente Principal (PC2)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Findings

* Las distribuciones univariadas mostraron patrones claros en SubTotal, IVA, Total y Peso Kgs mediante histogramas y KDE.

* Los boxplots revelaron outliers importantes en SubTotal, Total, Peso Kgs y Diferencia.

* Se observaron variaciones significativas de Total por Estatus de Viaje y de SubTotal por Sucursal.

* Los scatter plots confirmaron fuertes relaciones lineales entre SubTotal–Total y SubTotal–IVA, como esperado.

* La relación entre Peso Kgs y Diferencia mostró variabilidad relevante que podría requerir revisión.

* El análisis de correlación evidenció correlaciones muy fuertes entre variables financieras (SubTotal, Total, IVA).

* El PCA demostró que PC1 y PC2 explican el 100% de la varianza de SubTotal, Total, IVA y Peso Kgs.

* PC1 explica 75.26% de la varianza y PC2 el 24.74%, permitiendo reducir 4 variables a 2 sin pérdida de información.
